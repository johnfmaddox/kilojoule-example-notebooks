## Problem Statement

A piston-cylinder filled with oxygen, $\ce{O2}$, is heated. Determine the number of moles that will dissociate into monatomic oxygen, $\ce{O}$, by findin the minimum Gibbs function for the system when heated to $3000\ \mathrm{K}$ and $3200\ \mathrm{K}$.

## Chemical Balance

\begin{equation}
	\ce{ O2 -> \underbrace{x O2}_{\mathrm{leftover\\reactants}} + \underbrace{2(1-x) O}_{\mathrm{products}} }
\end{equation}

Number of moles:
\begin{equation}
	N_{tot} = x+2-2x = 2-x
\end{equation}
Mole fractions:
\begin{gather}
	y_{\ce{O2}} = \frac{x}{2-x} \qquad 
	y_{\ce{O}} = \frac{2-2x}{2-x}
\end{gather}

Equilibrium occurs when the Gibbs function is minimized
\begin{align}
	G&=N \overline{g}\\
	&=N_{tot}(y_{O2}\overline{g}_{\ce{O2}}+y_{O2}\overline{g}_{\ce{O2}}) \leftarrow \mathrm{Minimize}
\end{align}

### Python Initialization
We'll start by importing the libraries we will use for our analysis.  

```python
from kilojoule.templates.default import *
from kilojoule.tables import Cengel

O2 = A19 = Cengel.Table("A19")
O = A24 = Cengel.Table("A24")
A26 = Cengel.Table("A26")
A27 = Cengel.Table("A27")

# Pressure
p = Quantity(1,'atm')

# Universal Gas Constant
Rbar = Quantity(8.3144626,'J/K/mol')

# Define some LaTeX variable name substitutions
set_latex({
    'ho_molar_O':r'{\overline{h}^{\,\circ}}_{\mathrm{O}}',
    'ho_f_molar_O':r'{\overline{h}_f^{\,\circ}}_{\mathrm{O}}',
    'h_molar_O':r'{\overline{h}}_{\mathrm{O}}',
    'so_molar_O':r'{\overline{s}^{\,\circ}}_{\mathrm{O}}',
    'ho_molar_O2':r'{\overline{h}^{\,\circ}}_{\mathrm{O_2}}',
    'ho_f_molar_O2':r'{\overline{h}_f^{\,\circ}}_{\mathrm{O_2}}',
    'h_molar_O2':r'{\overline{h}}_{\mathrm{O_2}}',
    'so_molar_O2':r'{\overline{s}^{\,\circ}}_{\mathrm{O_2}}',
})

Summary();
```

```python
%%showcalc
"### Properties"
"Temperatures"
T_0 = Quantity(25,'degC')
T_1 = Quantity(3000,'K')

"Diatomic oxygen: $\mathrm{O_2}$"
ho_f_molar_O2 = A26.ho_f_molar('O2')
ho_molar_O2 = O2.h_molar(T_0)
h_molar_O2 = O2.h_molar(T_1)
so_molar_O2 = O2.so_molar(T_1)

"Monatomic oxygen: $\mathrm{O}$"
ho_f_molar_O = A26.ho_f_molar('O')
ho_molar_O = O.h_molar(T_0)
h_molar_O = O.h_molar(T_1)
so_molar_O = O.so_molar(T_1)
```

```python
%%showcalc
x = np.linspace(0.001,.999,100)
gbar = x*(ho_f_molar_O2 + h_molar_O2 - ho_molar_O2 - T_1*so_molar_O2 + T_1*Rbar*ln((x/(2-x)))) + (2-2*x)*(ho_f_molar_O + h_molar_O - ho_molar_O - T_1*so_molar_O + T_1*Rbar*ln((2-2*x)/(2-x)))
```

We can plot the variation of the Gibbs function with dissociation percentage and look for the value that minimizes the Gibbs function.

```python
plt.plot(x,gbar)
plt.title('Gibbs  variation with dissociation percentage for $\mathrm{O_2}\\to\mathrm{O}$ at 3000 K')
plt.xlabel(r'$x$')
plt.ylabel(r'$\overline{g}\ [\mathrm{kJ/kmol}]$');
```

The minimum value occurs between 90-100% of the products remaining in $\ce{O2}$ form (<10% dissociation).  We can replot to zoom in on the minimum.

```python
x = np.linspace(0.90,.99,100)
gbar = x*(ho_f_molar_O2 + h_molar_O2 - ho_molar_O2 - T_1*so_molar_O2 + T_1*Rbar*ln((x/(2-x)))) + (2-2*x)*(ho_f_molar_O + h_molar_O - ho_molar_O - T_1*so_molar_O + T_1*Rbar*ln((2-2*x)/(2-x)))
plt.plot(x,gbar)
plt.title('Gibbs  variation with dissociation percentage for $\mathrm{O_2}\\to\mathrm{O}$ at 3000 K')
plt.xlabel(r'$x$')
plt.ylabel(r'$\overline{g}\ [\mathrm{kJ/kmol}]$');
```

## Repeat at 3200 K

```python
%%showcalc
T_2 = Quantity(3200,'K')
h_molar_O2 = O2.h_molar(T_2)
so_molar_O2 = O2.so_molar(T_2)
h_molar_O = O.h_molar(T_2)
so_molar_O = O.so_molar(T_2)
```

```python
x = np.linspace(0.85,.95,100)
gbar = x*(ho_f_molar_O2 + h_molar_O2 - ho_molar_O2 - T_2*so_molar_O2 + T_2*Rbar*ln((x/(2-x)))) + (2-2*x)*(ho_f_molar_O + h_molar_O - ho_molar_O - T_2*so_molar_O + T_2*Rbar*ln((2-2*x)/(2-x)))
plt.plot(x,gbar)
plt.title('Gibbs  variation with dissociation percentage for $\mathrm{O_2}\\to\mathrm{O}$ at 3200 K')
plt.xlabel(r'$x$')
plt.ylabel(r'$\overline{g}\ [\mathrm{kJ/kmol}]$');
```

From this we see that the dissociation of $\ce{O2}\to\ce{O}$ occurs at a relatively higher percentage at 3200 K ($\approx 11\%$) than at the lower temperature of 3000 K ($\approx 5.5\%$)
